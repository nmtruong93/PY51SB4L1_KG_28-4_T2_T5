```python
repeat_patients = df.groupby(by = ['patient_id']).size().sort_values(ascending=False)
repeat_patients[repeat_patients >= 2].to_frame().reset_index()
```

```python
filtered_patients = repeat_patients[repeat_patients >=2] 
filtered_patients.to_frame().reset_index()

filtered_df = df[~df.patient_id.isin(filtered_patients.index)]
filtered_df
```

```python
df[['patient_id','cell_size_uniformity','cell_shape_uniformity']]
```

```python
def KQ(x):
    if ((x['cell_size_uniformity'] > 5) &
        x['cell_shape_uniformity'] > 5):
        return ('normal')
    else:
        return ('abnormal')
```

```python
df['KQ'] = df.apply(lambda x: KQ(x), axis = 1)
df[['patient_id', 'doctor_name', 'KQ']]
```
